```python
import numpy as np
import matplotlib as plt
import json
from sklearn.metrics import ConfusionMatrixDisplay
```

# Func»õii

```python
def pretty_print_stats(stats):
    print(f"ROC_AUC: {stats['roc_auc']}")
    print(f"Accuracy: {stats['accuracy']}")
    
    print('\nBENIGN TRAFFIC')
    benign = stats['benign']
    print(f"Precision: {benign['precision']}")
    print(f"Recall: {benign['recall']}")
    print(f"F1: {benign['f1']}")
    
    print('\nMALICIOUS TRAFFIC')
    mal = stats['attack']
    print(f"Precision: {mal['precision']}")
    print(f"Recall: {mal['recall']}")
    print(f"F1: {mal['f1']}")
    
    print("\n")
    print(f"False alerts: {stats['false_alerts']}")
    print(f"% malicious passed: {stats['attack_passed']:.4f}%")
```

```python
def confusion_matrix(stats):
    cm = np.array([ [stats['tn'], stats['fp']], [stats['fn'], stats['tp']] ])
    return ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.array(['Malicious', 'Benign']))
```

```python
ae_all = json.load(open('autoencoder_all.json', 'r'))
# print(json.dumps(ae_all, indent=4))

pretty_print_stats(ae_all)

cm = confusion_matrix(ae_all)
cm.plot(values_format='')
cm.figure_.savefig("ae_all_cm.png")
```
